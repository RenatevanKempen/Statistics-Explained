## Assignment:
Calculate the mean for grouped data of the dataframe 'df'

The formula for this --> mean = $\bar y$  $\sim$ $\frac{\sum_i f_i y_i} n$

where: 
- $y_i$ = midpoint of the $i$th class interval
- $n$ = total number of measurements
- $f_i$ = frequency associated with the $i$th class interval
- $\sim$ denotes 'is approximately equal to'

## Solution 

```python
#first let's add a total row to the dataframe: 
df.loc['Total']= df.sum()
df
```

With this information, we can fill in the formula

- $n$ = 100, so: 

$\bar y$  $\sim$ $\frac{\sum_i f_i y_i} {100}$

mean = $\bar y$  $\sim$ $\frac{2887.5} {100}$ = 28.875
